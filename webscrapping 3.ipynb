```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Scrape top 10 ODI teams
def scrape_odi_teams():
    url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table")
    rows = table.find_all("tr")

    team_data = []
    for row in rows[1:11]:
        cells = row.find_all("td")
        team_rank = cells[0].get_text().strip()
        team_name = cells[1].find("span", class_="u-hide-phablet").get_text().strip()
        matches = cells[2].get_text().strip()
        points = cells[3].get_text().strip()
        rating = cells[4].get_text().strip()

        team_data.append([team_rank, team_name, matches, points, rating])

    return team_data


# Scrape top 10 ODI batsmen
def scrape_odi_batsmen():
    url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table")
    rows = table.find_all("tr")

    batsman_data = []
    for row in rows[1:11]:
        cells = row.find_all("td")
        batsman_rank = cells[0].get_text().strip()
        batsman_name = cells[1].find("span", class_="u-hide-phablet").get_text().strip()
        batsman_team = cells[1].find("span", class_="table-body__logo-text").get_text().strip()
        batsman_rating = cells[2].get_text().strip()

        batsman_data.append([batsman_rank, batsman_name, batsman_team, batsman_rating])

    return batsman_data


# Scrape top 10 ODI bowlers
def scrape_odi_bowlers():
    url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table")
    rows = table.find_all("tr")

    bowler_data = []
    for row in rows[1:11]:
        cells = row.find_all("td")
        bowler_rank = cells[0].get_text().strip()
        bowler_name = cells[1].find("span", class_="u-hide-phablet").get_text().strip()
        bowler_team = cells[1].find("span", class_="table-body__logo-text").get_text().strip()
        bowler_rating = cells[2].get_text().strip()

        bowler_data.append([bowler_rank, bowler_name, bowler_team, bowler_rating])

    return bowler_data


# Create dataframes for each category
team_data = pd.DataFrame(scrape_odi_teams(), columns=["Rank", "Team", "Matches", "Points", "Rating"])
batsman_data = pd.DataFrame(scrape_odi_batsmen(), columns=["Rank", "Batsman", "Team", "Rating"])
bowler_data = pd.DataFrame(scrape_odi_bowlers(), columns=["Rank", "Bowler", "Team", "Rating"])

# Display the dataframes
print("Top 10 ODI Teams:")
print(team_data)
print("\nTop 10 ODI Batsmen:")
print(batsman_data)
print("\nTop 10 ODI Bowlers:")
print(bowler_data)
```
