```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Scrape top 10 ODI teams in women's cricket
def scrape_odi_teams_women():
    url = "https://www.icc-cricket.com/rankings/womens/team-rankings/odi"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table")
    rows = table.find_all("tr")

    team_data = []
    for row in rows[1:11]:
        cells = row.find_all("td")
        team_rank = cells[0].get_text().strip()
        team_name = cells[1].find("span", class_="u-hide-phablet").get_text().strip()
        matches = cells[2].get_text().strip()
        points = cells[3].get_text().strip()
        rating = cells[4].get_text().strip()

        team_data.append([team_rank, team_name, matches, points, rating])

    return team_data


# Scrape top 10 women's ODI batting players
def scrape_odi_batting_players_women():
    url = "https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table")
    rows = table.find_all("tr")

    batting_data = []
    for row in rows[1:11]:
        cells = row.find_all("td")
        player_rank = cells[0].get_text().strip()
        player_name = cells[1].find("span", class_="u-hide-phablet").get_text().strip()
        player_team = cells[1].find("span", class_="table-body__logo-text").get_text().strip()
        player_rating = cells[2].get_text().strip()

        batting_data.append([player_rank, player_name, player_team, player_rating])

    return batting_data


# Scrape top 10 women's ODI all-rounders
def scrape_odi_allrounders_women():
    url = "https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table")
    rows = table.find_all("tr")

    allrounder_data = []
    for row in rows[1:11]:
        cells = row.find_all("td")
        player_rank = cells[0].get_text().strip()
        player_name = cells[1].find("span", class_="u-hide-phablet").get_text().strip()
        player_team = cells[1].find("span", class_="table-body__logo-text").get_text().strip()
        player_rating = cells[2].get_text().strip()

        allrounder_data.append([player_rank, player_name, player_team, player_rating])

    return allrounder_data


team_data_women = pd.DataFrame(scrape_odi_teams_women(), columns=["Rank", "Team", "Matches", "Points", "Rating"])
batting_data_women = pd.DataFrame(scrape_odi_batting_players_women(), columns=["Rank", "Player", "Team", "Rating"])
allrounder_data_women = pd.DataFrame(scrape_odi_allrounders_women(), columns=["Rank", "Player", "Team", "Rating"])

# Display the dataframes
print("Top 10 ODI Teams in Women's Cricket:")
print(team_data_women)
print("\nTop 10 Women's ODI Batting Players:")
print(batting_data_women)
print("\nTop 10 Women's ODI All-rounders:")
print(allrounder_data_women)
```
